## Previo de la Práctica 1

### Definición de URDF.
El **URDF (Unified Robot Description Format)** es un formato XML utilizado para describir modelos de robots. Este formato permite definir las características físicas y cinemáticas de un robot, incluyendo sus eslabones (links) y juntas (joints). Es comúnmente utilizado en herramientas del Sistema Operativo de Robots (ROS), como RViz y el simulador Gazebo.

### Tipos de juntas que están definidas dentro del formato URDF.
Dentro del formato URDF, se pueden definir varios tipos de juntas, cada una con características específicas:
- **Revolute**: Permite la rotación alrededor de un eje con un rango limitado.
- **Continuous**: Permite la rotación continua alrededor de un eje sin límites.
- **Prismatic**: Permite el deslizamiento a lo largo de un eje con un rango limitado.
- **Fixed**: No permite ningún tipo de movimiento, es una conexión rígida.
- **Floating**: Permite el movimiento en los seis grados de libertad.
- **Planar**: Permite el movimiento en un plano perpendicular al eje.

### Definición de elementos "primitivos" que se pueden utilizar para definir un robot en ROS.
Para definir un robot en ROS, se utilizan varios elementos "primitivos" en el archivo URDF:
- **Links**: Representan las partes físicas del robot.
- **Joints**: Describen cómo se conectan y se mueven los enlaces entre sí.
- **Visual**: Define la apariencia visual del robot.
- **Collision**: Define el modelo de colisión del robot.
- **Inertial**: Define las propiedades físicas como la masa y el centro de masa.

### ¿Qué es un nodo y que es un topico en ROS 2?
- **Nodo**: En ROS 2, un nodo es una entidad básica que ejecuta una tarea en el sistema. Cada nodo puede publicar, suscribirse, recibir y enviar mensajes a otros nodos a través de tópicos. Los nodos están diseñados para ser modulares, de modo que cada componente de un sistema robótico complejo pueda funcionar de forma independiente.

- **Tópico**: Un tópico es un canal de comunicación a través del cual los nodos se comunican. Los nodos pueden publicar datos en un tópico o suscribirse a un tópico para recibir datos. Esta arquitectura de comunicación basada en tópicos es clave para la modularidad y escalabilidad de ROS.


### ¿Qué es RVIZ?
**RVIZ** es una herramienta de visualización en ROS que permite ver el estado y la configuración de un robot en tiempo real. Se utiliza para depurar y analizar el comportamiento de los robots, mostrando datos como la posición, orientación, y sensores del robot.


## Desarrollo

### Parte 1
Desarrollar y simular la descripción de un eslabón cuyo movimiento esté restringido por una junta rotacional.

***Código generado en clase***  

``` xml
<?xml version="1.0" ?>

<robot name="basic_link">

    <link name="base_link">
        <visual>
            <origin xyz="0.0 0.0 0.10" rpy="0.0 0.0 0.0" />
            <geometry>
                <box size="0.15 0.10 0.20" />
            </geometry>
            <material name="AzulClaro">
                <color rgba="0.678 0.847 0.902 1.0" />
            </material>
        </visual>

        <collision>
            <origin xyz="0.0 0.0 0.10" rpy="0.0 0.0 0.0" />
            <geometry>
                <box size="0.15 0.10 0.20" />
            </geometry>
        </collision>

        <inertial>
            <mass value="1.0" />
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0" />
            <inertia ixx="1.0" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0" />
        </inertial>
    </link>

    <link name="link_1">
        <visual>
            <origin xyz="0.0 0.05 0.25" rpy="0.0 0.0 0.0" />
            <geometry>
                <box size="0.1 0.1 0.5" />
            </geometry>
            <material name="Coral">
                <color rgba="1.0 0.498 0.314 1.0" />
            </material>
        </visual>

        <collision>
            <origin xyz="0.0 0.05 0.25" rpy="0.0 0.0 0.0" />
            <geometry>
                <box size="0.1 0.1 0.5" />
            </geometry>
        </collision>

        <inertial>
            <mass value="1.0" />
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0" />
            <inertia ixx="1.0" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0" />
        </inertial>
    </link>

    <joint name="base_link_to_link_1_joint" type="revolute">
        <parent link="base_link" />
        <child link="link_1" />
        <axis xyz="0 1 0" />
        <limit lower="-1.57" upper="1.57" velocity="1" 
            effort="1000" />
        <origin xyz="0.0 0.05 0.20" rpy="0 0 0" />
    </joint>

</robot>
```

Este código URDF define un robot simple con dos eslabones (links) llamados “base_link” y “link_1”, el "base_link" estará fijado a tierra, ambos eslabones están conectados por una junta de revolución (revolute joint). Cada eslabón tiene propiedades visuales, de colisión e inerciales. 
La junta rotacional, llamada “base_link_to_link_1_joint”, conecta “base_link” con “link_1” y permite la rotación alrededor del eje Y. Esta junta tiene límites de rotación entre -1.57 y 1.57 radianes, una velocidad máxima de 1 rad/s y un esfuerzo máximo de 1000 N·m.


### Parte 2
Modificar el código del archivo para agregar otro eslabón el cual puede estar unido por al primer eslabón por una junta revolución o una junta prismática.

***Fragmento del código dónde se agrega un eslabón y una junta*** [basic_link.urdf](https://github.com/DavidJara14/robotica_dpj-2025-1/blob/main/practica1_description/urdf/basic_link.urdf)

``` xml
    <link name="link_2">
        <visual>
            <origin xyz="0.0 0.05 0.2" rpy="0.0 0.0 0.0" />
            <geometry>
                <box size="0.1 0.1 0.4" />
            </geometry>
            <material name="GrisSuave">
                <color rgba="0.827 0.827 0.827 1.0" />
            </material>
        </visual>

        <collision>
            <origin xyz="0.0 0.05 0.2" rpy="0.0 0.0 0.0" />
            <geometry>
                <box size="0.1 0.1 0.4" />
            </geometry>
        </collision>

        <inertial>
            <mass value="1.0" />
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0" />
            <inertia ixx="1.0" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0" />
        </inertial>
    </link>

    <joint name="link_1_to_link_2_joint" type="revolute">
        <parent link="link_1" />
        <child link="link_2" />
        <axis xyz="0 1 0" />
        <limit lower="-1.57" upper="1.57" velocity="1" 
            effort="1000" />
        <origin xyz="0.0 0.10 0.5" rpy="0 0 0" />
    </joint>
```

Este código urdf añade el eslabón "link_2", con las mismas propiedades visuales, de colisión e inerciales, y crea una junta rotacional denominada "link_1_to_link_2_joint" con las mismas caracteriasticas que la definida arriba.


Es importante mencionar que se creo el archivo launch [basic_link_display.launch.xml](https://github.com/DavidJara14/robotica_dpj-2025-1/blob/main/practica1_description/launch/basic_link_display.launch.xml), en el cual se configuran los nodos necesarios para visualizar un robot en RVIZ, se definen las rutas a los archivos urdf y de configuración de RVIZ, se inicia el "robot_state_publisher" y el "joint_state_publisher_gui" para publicar el estado y  manipular las articulaciones del robot respectivamente, por último se inicia RVIZ para visualizar el robot.
En el directorio rviz se creo el archivo para guardar las configuraciones del programa de rviz [basic_link_display.rviz](https://github.com/DavidJara14/robotica_dpj-2025-1/blob/main/practica1_description/rviz/basic_link_display.rviz)
